Part I

```python
import numpy as np
import pandas as pd
import scipy.stats as stats

# Raw data from Table 10.6
males = [12, 7, 7, 10, 8, 10, 11, 9, 9, 13,
         4, 9, 12, 11, 9, 9, 7, 12, 10, 13,
         11, 10, 6, 12, 11, 9, 10, 12, 8, 9,
         13, 10, 9, 7, 10, 7, 10, 8, 11, 10,
         11, 7, 15, 8, 9, 9, 11, 13, 10, 13]

females = [11, 10, 11, 10, 11, 12, 12, 10, 9, 9,
           9, 10, 8, 7, 12, 9, 7, 8, 9, 8,
           7, 7, 9, 9, 12, 10, 9, 13, 9, 9,
           10, 9, 6, 12, 8, 11, 8, 8, 11, 12,
           9, 10, 11, 14, 12, 7, 11, 10, 9, 11]

# Combine data into a DataFrame
df = pd.DataFrame({
    'Hours': males + females,
    'Gender': ['Male']*50 + ['Female']*50
})

def compute_stats(data):
    n = len(data)
    mean = np.mean(data)
    std_err = stats.sem(data)
    ci = stats.t.interval(0.95, df=n-1, loc=mean, scale=std_err)

    return {
        'N': n,
        'Mean': mean,
        'Median': np.median(data),
        'SD': np.std(data, ddof=1),
        'Variance': np.var(data, ddof=1),
        'Min': np.min(data),
        'Max': np.max(data),
        'Skewness': stats.skew(data, bias=False),
        'Kurtosis': stats.kurtosis(data, bias=False),
        'Q1 (25%)': np.percentile(data, 25),
        'Q3 (75%)': np.percentile(data, 75),
        '95% CI Lower': ci[0],
        '95% CI Upper': ci[1]
    }

# Execute computations
overall_stats = compute_stats(df['Hours'])
male_stats = compute_stats(df[df['Gender'] == 'Male']['Hours'])
female_stats = compute_stats(df[df['Gender'] == 'Female']['Hours'])
```

```python
import numpy as np
import scipy.stats as stats

# ---------- Part II: Moments ----------
x = np.array([2, 3, 7, 8, 10, 10, 12, 15, 17, 20])

# 1. Raw moments (moments about zero, denoted as m'_k)
m1_raw = np.mean(x)
m2_raw = np.mean(x**2)
m3_raw = np.mean(x**3)
m4_raw = np.mean(x**4)

# 2. Central moments (moments about the mean, denoted as m_k)
# The first central moment is mathematically always 0.
m1_cent = np.mean(x - m1_raw)
m2_cent = np.mean((x - m1_raw)**2)
m3_cent = np.mean((x - m1_raw)**3)
m4_cent = np.mean((x - m1_raw)**4)

# 3. Verification of the formula
m4_formula = m4_raw - 4*m1_raw*m3_raw + 6*(m1_raw**2)*m2_raw - 3*(m1_raw**4)

print("--- Part II Answers ---")
print(f"1. Raw Moments: m'_1={m1_raw}, m'_2={m2_raw}, m'_3={m3_raw}, m'_4={m4_raw}")
print(f"2. Central Moments: m_1={m1_cent:.2f}, m_2={m2_cent}, m_3={m3_cent}, m_4={m4_cent}")
print(f"3. Verification: Left Side = {m4_cent}, Right Side = {m4_formula}")


# ---------- Part IV: Student's t-distribution ----------
print("\n--- Part IV Answers ---")
# 1. Area between -t1 and t1 is 0.90, v = 20
# Total tail area = 0.10. One tail = 0.05. Percentile = 0.95
t1_1 = stats.t.ppf(0.95, df=20)
print(f"1. t1 = {t1_1:.3f}")

# 2. Area to the left of -t1 is 0.025, v = 20
# -t1 is the 2.5th percentile, making t1 the 97.5th percentile.
t1_2 = stats.t.ppf(0.975, df=20)
print(f"2. t1 = {t1_2:.3f}")

# 3. Combined area to the right of t1 and left of -t1 is 0.01, v = 5
# Total tail area = 0.01. One tail = 0.005. Percentile = 0.995
t1_3 = stats.t.ppf(0.995, df=5)
print(f"3. t1 = {t1_3:.3f}")

# 4. Area to the right of t1 is 0.55, v = 16
# Area to the left is 1 - 0.55 = 0.45.
t1_4 = stats.t.ppf(0.45, df=16)
print(f"4. t1 = {t1_4:.3f}")
```

Prove that $m'_4 = m_4 + 4hm_3 + 6h^2m_2 + h^4$, where $h = m'_1$.

By definition, the $k$-th raw moment is $m'_k = \frac{1}{n}\sum x^k$, and the $k$-th central moment is $m_k = \frac{1}{n}\sum (x - h)^k$.
Because $h$ represents the mean ($m'_1 = \bar{x}$), we can rewrite the individual values $x$ in the equation for the 4th raw moment by adding and subtracting $h$:

$$
\begin{align*}
m'_4 &= \frac{1}{n} \sum x^4 \\
     &= \frac{1}{n} \sum ((x - h) + h)^4
\end{align*}
$$

Using the binomial expansion theorem for $(a + b)^4 = a^4 + 4a^3b + 6a^2b^2 + 4ab^3 + b^4$, we expand the term inside the summation where $a = (x-h)$ and $b = h$:

$$
m'_4 = \frac{1}{n} \sum \left[ (x - h)^4 + 4(x - h)^3h + 6(x - h)^2h^2 + 4(x - h)h^3 + h^4 \right]
$$

Now, distribute the summation operator $\left(\frac{1}{n}\sum \right)$ across the terms:

$$
m'_4 = \frac{1}{n}\sum(x - h)^4 + 4h\left[\frac{1}{n}\sum(x - h)^3\right] + 6h^2\left[\frac{1}{n}\sum(x - h)^2\right] + 4h^3\left[\frac{1}{n}\sum(x - h)\right] + \frac{1}{n}\sum h^4
$$

Next, substitute the definitions of the central moments ($m_k$) back into the equation:
1. $\frac{1}{n}\sum(x - h)^4$ is $m_4$
2. $\frac{1}{n}\sum(x - h)^3$ is $m_3$
3. $\frac{1}{n}\sum(x - h)^2$ is $m_2$
4. $\frac{1}{n}\sum(x - h)$ is $m_1$, which is always equal to **0** because the sum of deviations from the mean is exactly zero.
5. $\frac{1}{n}\sum h^4$ simplifies to $h^4$ (because the average of a constant is the constant itself).

Substituting these yields:

$$
\begin{align*}
m'_4 &= m_4 + 4h(m_3) + 6h^2(m_2) + 4h^3(0) + h^4 \\
m'_4 &= m_4 + 4hm_3 + 6h^2m_2 + h^4
\end{align*}
