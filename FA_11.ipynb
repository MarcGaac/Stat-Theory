```r
# Install ggplot2 if you haven't already
# install.packages("ggplot2")

library(ggplot2)

# Create a publication-ready boxplot
ggplot(PlantGrowth, aes(x = group, y = weight, fill = group)) +
  geom_boxplot(alpha = 0.7) +
  scale_fill_manual(values = c("#E0E0E0", "#A9CCE3", "#A3E4D7")) +
  labs(x = "Treatment Group",
       y = "Dried Weight (g)") +
  scale_x_discrete(labels = c("ctrl" = "Control", "trt1" = "Treatment 1", "trt2" = "Treatment 2")) +
  theme_classic() +
  theme(legend.position = "none",
        text = element_text(size = 12))
```

```r
# 1. Run the one-way ANOVA
anova_model <- aov(weight ~ group, data = PlantGrowth)

# 2. View the main ANOVA table (shows the F-statistic and p-value)
summary(anova_model)

# 3. Run the Tukey HSD post-hoc test (to see which specific groups differ)
TukeyHSD(anova_model)
```

**Method**

**Data Description**

Thirty observations were taken on the current dataset, called PlantGrowth, which includes the dried weights (grams) of plants that have been grown under one of three different conditions; control (ctrl), treatment 1 (trt1), and treatment 2 (trt2). Each condition contained 10 individual plants.

**Statistical Analysis**

A one-way analysis of variance (ANOVA) was performed to determine if treatment conditions revealed significant differences when compared using the null hypothesis which states that there is no significant correlation between the variables being analyzed. Prior to performing an ANOVA test, the assumptions regarding normality and variance were checked using the following two tests: Shapiro-Wilk Test was used to check if the residuals’ distributions were normal, and Levene's Test was used to confirm the homogeneity of variances. A post hoc test (Tukey's HSD) was then used for statistically significant level (α = .05) main effects as a means of controlling for Type I error rates with multiple pairwise comparisons.

**Results**

**Assumption Testing**

The normal distribution of the ANOVA residuals fulfilled the Shapiro-Wilk test criteria, W = 0.966, p = .438. In addition, Levene's test, F(2, 27) = 1.119, p = .341, demonstrated that the variances of the three groups from which the ANOVA is drawn are homogeneous. As both assumptions were satisfied, no data transformations were needed and the standard one-way ANOVA was performed.

**Descriptive Statistics**

Based on the descriptive statistics, the heaviest plants were in Treatment 2 (M = 5.53, SD = 0.44), whereas the Control group was next (M = 5.03, SD = 0.58), followed by Treatment 1 (M = 4.66, SD = 0.79), which had the lightest plants.

**Primary ANOVA Results**

The one-way ANOVA revealed a statistically significant difference in plant weights between the three groups, F(2, 27) = 4.85, p = .016, $\eta^2$ = .26. The effect size ($\eta^2$ = .26) indicates a large effect, meaning approximately 26.4% of the variance in plant weight is accounted for by the treatment conditions. Because the p-value is less than .05, we reject the null hypothesis that there is no significant difference in weight between the treatment groups.

**Post-Hoc Analysis**

In the analysis of the data, a Tukey HSD post-hoc test produced a statistically significant difference between the mean weight of plants in Treatment 1 and Treatment 2 (Mean Difference = 0.865, p = .012). There were no statistically significant differences in plant weight between Control and Treatment 1 (Mean Difference = -0.371, p = .391) nor between Control and Treatment 2 (Mean Difference = 0.494, p = .198).

**Discussion**

The purpose of this analysis was to determine whether if the different treatments significantly affected plant growth (measured by dried weight). The results which has lead us to reject the null hypothesis, as a significant main effect of the treatment group was observed. Specifically, Treatment 2 was shown to be significantly more effective at promoting plant growth compared to Treatment 1. However, neither of the treatment produced a statistically significant difference in weight when directly compared to the control condition. For future research with a larger sample size may be beneficial to determine if the moderate mean differences seen between the control group and the two treatments represent true population effects that lacked statistical power in this 30-plant sample.
